

This code loads a dataset about Netflix titles, prints the first few rows, and then creates two visualizations: a bar chart showing the distribution of movies and TV shows and a histogram showing the distribution of release years.

```python


import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
netflix_data = pd.read_csv("C:\\netflix_titles.csv" , encoding='latin1')

# Display the first few rows of the dataset to understand its structure
print(netflix_data.head())

# Visualize the distribution of movies and TV shows
plt.figure(figsize=(10, 5))

# Plot bar chart for type distribution
plt.subplot(1, 2, 1)
netflix_data['type'].value_counts().plot(kind='bar', color=['blue', 'green'])
plt.title('Distribution of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0) 

# Plot histogram for release year distribution
plt.subplot(1, 2, 2)
plt.hist(netflix_data['release_year'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
